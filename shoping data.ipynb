```python
from sklearn.cluster import KMeans
algo=KMeans(n_clusters=2)
algo.fit(data)
```

```python
cen=algo.cluster_centers_
```

```python
algo.labels_
```

```python
algo.inertia_
```

```python
sns.scatterplot(data['Annual Income (k$)'],data['Spending Score (1-100)'],hue=algo.labels_)
sns.scatterplot(cen[:,0],cen[:,1],color='r')
```

```python
dis=[]
k=range(1,10)
for i in k:
    algo=KMeans(n_clusters=i)
    algo.fit(data)
    dis.append(algo.inertia_)
```

```python
dis
```

```python
plt.plot(k,dis)
plt.grid()
plt.show()
```

```python
algo1=KMeans(n_clusters=5)
algo1.fit(data)
```

```python
cen=algo1.cluster_centers_
cen
```

```python
sns.scatterplot(data['Annual Income (k$)'],data['Spending Score (1-100)'],hue=algo1.labels_)
sns.scatterplot(cen[:,0],cen[:,1],color='b')
```

In cluster 1(center of the graph ) we see that people have high income and high spending scores, this is the ideal case for the mall or shops as these people are the prime sources of profit. These people might be the regular customers of the mall and are convinced by the mall’s facilities
